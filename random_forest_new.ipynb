```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cdtx_train_X = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/cdtx_train_X.csv')
cdtx_train_y = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/cdtx_train_y.csv')
cdtx_train_idx = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/cdtx_train_idx.csv')

dp_train_X = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/dp_train_X.csv')
dp_train_y = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/dp_train_y.csv')
dp_train_idx = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/dp_train_idx.csv')

remit_train_X = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/remit_train_X.csv')
remit_train_y = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/remit_train_y.csv')
remit_train_idx = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/remit_train_idx.csv')

cdtx_test_X = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/cdtx_test_X.csv')
cdtx_test_idx = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/cdtx_test_idx.csv')

dp_test_X = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/dp_test_X.csv')
dp_test_idx = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/dp_test_idx.csv')

remit_test_X = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/remit_test_X.csv')
remit_test_idx = pd.read_csv('/home/bill/moneylaundring/contest/processed_dataset/remit_test_idx.csv')
```

```python
cdtx_train_X
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
```

```python
dp_test_X
```

```python
cdtx_ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(handle_unknown='ignore'),[0,1,3,5,6,7])], remainder='passthrough')
cdtx_X_train = cdtx_ct.fit_transform(cdtx_train_X)
cdtx_X_test = cdtx_ct.transform(cdtx_test_X)
```

```python
dp_ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(handle_unknown='ignore'),[0,1,5,7,8,9,10,11,12])], remainder='passthrough')
dp_X_train = dp_ct.fit_transform(dp_train_X)
dp_X_test = dp_ct.transform(dp_test_X)
```

```python
r_ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(handle_unknown='ignore'),[0,1,5,7])], remainder='passthrough')
remit_X_train = r_ct.fit_transform(remit_train_X)
remit_X_test = r_ct.transform(remit_test_X)
```

```python
cdtx_X_train.shape
```

```python
regressor_cdtx = RandomForestRegressor(n_estimators=100, criterion='squared_error', random_state= 42)
regressor_cdtx.fit(cdtx_X_train, cdtx_train_y.values.ravel())
```

```python
dp_X_train.shape
```

```python
regressor_dp = RandomForestRegressor(n_estimators=100, criterion='squared_error', random_state= 42)
regressor_dp.fit(dp_X_train, dp_train_y.values.ravel())
```

```python
regressor_remit = RandomForestRegressor(n_estimators=100, criterion='squared_error', random_state= 42)
regressor_remit.fit(remit_X_train, remit_train_y.values.ravel())
```

```python
cdtx_y_pred = regressor_cdtx.predict(cdtx_X_test)
cdtx_y_pred
```

```python
cdtx_y_ans = pd.DataFrame(cdtx_y_pred)
cdtx_y_ans[cdtx_y_ans[0]!=0]
```

```python
dp_y_pred = regressor_dp.predict(dp_X_test)
dp_y_ans = pd.DataFrame(dp_y_pred)
dp_y_ans
```

```python
remit_y_pred = regressor_remit.predict(remit_X_test)
remit_y_ans = pd.DataFrame(remit_y_pred)
remit_y_ans
```

```python
cdtx_ans = pd.DataFrame(cdtx_test_idx['alert_key'])
cdtx_ans['probability'] = cdtx_y_ans[0]
cdtx_ans
```

```python
dp_ans = pd.DataFrame(dp_test_idx['alert_key'])
dp_ans['probability'] = dp_y_ans[0]
dp_ans
```

```python
remit_ans = pd.DataFrame(remit_test_idx['alert_key'])
remit_ans['probability'] = remit_y_ans[0]
remit_ans
```

```python
ans = pd.concat([cdtx_ans,dp_ans,remit_ans])
```

```python
ans
```

```python
ans['probability'] = ans['probability'].abs()
```

```python
ans_mean = ans.groupby(['alert_key']).max()
```

```python
ans_mean
```

```python
submit = pd.read_csv('/home/bill/moneylaundring/dataset/submit.csv')
```

```python
submit['probability']=0
```

```python
submit_new = pd.merge(submit, ans_mean, on=['alert_key'],how='left')
```

```python
submit_new['probability_y'].fillna(0,inplace=True)
```

```python
submit_final = submit_new.rename(columns={'probability_y':'probability'})
```

```python
submit_final = submit_final[['alert_key','probability']]
```

```python
submit_final = submit_final.sort_values(by ='probability', ascending=False)
```
